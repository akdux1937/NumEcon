# Static: Indifference curves and checks

Using **analytical** expressions:

```python
# code
consumer = consumption.ConsumerClass(preferences='cobb_douglas',alpha=0.1,beta=0.1,x1max=10,x2max=10)
fig,ax = consumer.figure()

# 45 degrees 
ax.plot([0,consumer.x1max],[0,consumer.x2max],'--',color="black",zorder=1,alpha=0.1)

# indifference curves
x1s = [2,4,6]
x2s = [2,4,6]
us = [consumer.utility(x1,x2) for x1,x2 in zip(x1s,x2s)]
[ax.plot(x1,x2,'ro',color='black') for x1,x2 in zip(x1s,x2s)]
[ax.text(x1*1.03,x2*1.03,f'u = {u:5.2f}') for x1,x2,u in zip(x1s,x2s,us)]
consumer.plot_indifference_curves(ax,us)

# checks
consumer.plot_monotonicity_check(ax,x1=x1s[1],x2=x2s[1])
consumer.plot_convexity_check(ax,u=us[0])
```

Using **numerical** solution:

```python
# code
consumer = consumption.ConsumerClass(preferences='cobb_douglas',alpha=0.1,beta=0.1,x1max=10,x2max=10)
fig,ax = consumer.figure()

# 45 degrees 
ax.plot([0,consumer.x1max],[0,consumer.x2max],'--',color="black",zorder=1,alpha=0.1)

# indifference curves
x1s = [2,4,6]
x2s = [2,4,6]
us = [consumer.utility(x1,x2) for x1,x2 in zip(x1s,x2s)]
[ax.plot(x1,x2,'ro',color='black') for x1,x2 in zip(x1s,x2s)]
[ax.text(x1*1.03,x2*1.03,f'u = {u:5.2f}') for x1,x2,u in zip(x1s,x2s,us)]
consumer.plot_indifference_curves(ax,us,numerical=True,only_dots=True)
```
